```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("dataset aluminium lme.xlsx")
df.head()
```

```python
dm = df.loc[0:6]
# dm = df
dm
```

```python
# for i in range(len(dm["LME Aluminium Cash"])):
#     if i == 0:
#         print ((dm["LME Aluminium Cash"][i:(i+1)*3]) )
#         print (max(dm["LME Aluminium Cash"][i:(i+1)*3]))
#         print (min(dm["LME Aluminium Cash"][i:(i+1)*3]))

#         print("target: ", dm["LME Aluminium Cash"][i+3])
#     elif (i)+3 < len(dm["LME Aluminium Cash"]):
#         print ((dm["LME Aluminium Cash"][i:(i)+3] ))
#         print("target: ", dm["LME Aluminium Cash"][i+3])
normalData = pd.DataFrame(columns= ["x1", "x2", "y"])
for i in range(len(dm["LME Aluminium Cash"])):
    if i == 0:
        normalData.loc[i, "x1"] = max(dm["LME Aluminium Cash"][i:(i+1)*3])
        normalData.loc[i, "x2"] = min(dm["LME Aluminium Cash"][i:(i+1)*3])
        normalData.loc[i, "y"] = dm["LME Aluminium Cash"][i+3]
    elif (i)+3 < len(dm["LME Aluminium Cash"]):
        normalData.loc[i, "x1"] = max(dm["LME Aluminium Cash"][i:(i)+3])
        normalData.loc[i, "x2"] = min(dm["LME Aluminium Cash"][i:(i)+3] )
        normalData.loc[i, "y"] = dm["LME Aluminium Cash"][i+3]
normalData
```

```python
def minmax(data):
    minimum = {}
    maximum = {}
    for i in data.columns:
        minimum[i] = data[i].min()
        maximum[i] = data[i].max()
    for i in data.index:
        for j in data.columns:
            data.loc[i, j] = (data.loc[i,j] - minimum[j]) / (maximum[j] - minimum[j])
    return data
minmax(normalData)
```

```python
def sigmoid(value):
    return 1 / (1 + np.exp(-value))
```

```python
def differ_sigmoid(value):
    return sigmoid(value) * (1 - sigmoid(value))
```

```python
b_in = 0.1
layer = 15
b_out = 0.1
alpha = 0.1
```

```python
weightArray = []
for i in range(2):
    weightArray.append(pd.DataFrame(index=[0]))
for i in range(len(weightArray)):
    for j in range(layer):
        weightArray[i].loc[0, j] = (j+1) * 0.1
weightArray[1]
```

```python
outputLayer = pd.DataFrame(columns=[i for i in range(layer)])
for i in range(len(outputLayer.columns)):
    outputLayer.loc[0, i] = (i+1) * 0.1
outputLayer
```

```python
z_net = pd.DataFrame(0, index =[i for i in range(len(normalData))], 
                     columns=[i for i in range(layer)])
for i in z_net.index:
    for j in z_net.columns:
        z_net.loc[i, j] = b_in + normalData.loc[i, "x1"] * weightArray[0].loc[0, j] + normalData.loc[i, "x2"] * weightArray[1].loc[0, j]
z_net
```

```python
z_activate = pd.DataFrame(0, index =[i for i in range(len(normalData))], 
                          columns=[i for i in range(layer)])
for i in z_activate.index:
    for j in z_activate.columns:
#         z_activate.loc[i, j] = 1 / (1+(2.71828183 ** -(z_net.loc[i, j])))
        z_activate.loc[i, j] = sigmoid(z_net.loc[i, j])

z_activate
```

```python
y_net = []
y_temp = []
for i in range(len(normalData)):
    y_temp.append([])
for i in range(len(normalData)):
    for j in (z_activate.columns):
        y_temp[i].append((z_activate.loc[i, j] * outputLayer.loc[0, j]))
    y_net.append(np.sum(y_temp[i])+b_out)
y_net
```

```python
y_activate = []
for i in y_net:
    y_activate.append(sigmoid(i))
y_activate
```

```python
backOutputTarget = []
for i in range(len(y_activate)):
    backOutputTarget.append((normalData.loc[i, "y"] - 
                             y_activate[i]) * differ_sigmoid(y_net[i]))
backOutputTarget            
```

```python
backAlphaHidden = pd.DataFrame(0, index = [i for i in range(len(normalData))], 
                               columns = [i for i in range(layer)])
for i in backAlphaHidden.index:
    for j in backAlphaHidden.columns:
        backAlphaHidden.loc[i, j] = alpha * backOutputTarget[i] * z_activate.loc[i, j]
backAlphaHidden
```

```python
errorTarget = pd.DataFrame(0, index = [i for i in range(len(normalData))], 
                               columns = [i for i in range(layer)])
for i in errorTarget.index:
    for j in errorTarget.columns:
        errorTarget.loc[i, j] = backOutputTarget[i] * outputLayer.loc[0, j]
errorTarget
```

```python
errorHidden = pd.DataFrame(0, index = [i for i in range(len(normalData))], 
                               columns = [i for i in range(layer)])
for i in errorHidden.index:
    for j in errorHidden.columns:
        errorHidden.loc[i, j] = z_activate.loc[i, j] * (1 - z_activate.loc[i, j]) * errorTarget.loc[i, j]
errorHidden
```

```python
outputWeightCorrection = pd.DataFrame(0, index = [i for i in range(len(normalData))], 
    columns = [i for i in range(layer)])
outputWeightCorrection["b"] = 0
for i in range(len(backOutputTarget)):
    outputWeightCorrection.loc[i, "b"] = b_out * backOutputTarget[i]
for i in z_activate.index:
    for j in z_activate.columns:
        outputWeightCorrection.loc[i, j] = b_out * z_activate.loc[i, j] * backOutputTarget[i]
outputWeightCorrection
```

```python
deltaInputHidden = []
for i in range(2):
    deltaInputHidden.append(pd.DataFrame(index=[i for i in range(len(normalData))]))
for i in range(len(deltaInputHidden)):
    for j in range(len(normalData)):
        for k in range(layer):
            deltaInputHidden[i].loc[j, k] = alpha * backAlphaHidden.loc[j, k] * normalData.loc[j, normalData.columns[i]]
deltaInputHidden[1]
```

```python
newWeightArray = []
for i in range(2):
    newWeightArray.append(pd.DataFrame(index=[0]))
for i in range(len(newWeightArray)):
    for j in range(len(normalData)):
        for k in range(layer):
            newWeightArray[i].loc[j, k] = deltaInputHidden[i].loc[j, k] + weightArray[i].loc[0, k]

newWeightArray[1]
```

```python
newOutputLayer = pd.DataFrame(index = [i for i in range(len(normalData))], columns=[i for i in range(layer)])
for i in newOutputLayer.index:
    for j in newOutputLayer.columns:
        newOutputLayer.loc[i, j] = outputWeightCorrection.loc[i, j] + outputLayer.loc[0, j]
for i in newOutputLayer.index:
    newOutputLayer.loc[i, "b"] = outputWeightCorrection.loc[i, "b"] + b_out
newOutputLayer
```

```python
from sklearn.metrics import mean_absolute_error 
from sklearn.metrics import mean_squared_error 
from math import sqrt

#MAE
mae = mean_absolute_error(normalData["y"], y_activate)
print('MAE: %f' % mae)

#MSE
mse = mean_squared_error(normalData["y"], y_activate)
print('MSE: %f' % mse)

#RMSE
rmse = sqrt(mse)
print('RMSE: %f' % rmse)

#MAPE (Mean Absolute Percentage Error)
actual = np.array([normalData["y"]])
forecast = np.array([y_activate])
mape=np.mean(np.abs((actual - forecast) / actual)) * 100
print('MAPE: %f' % mape)

# ACCURACY
Accuracy= 100-mape
print('Accuracy: %f' % Accuracy)
```

```python
EPSILON = 1e-10

def maape(actual, predicted):
    """
    Mean Arctangent Absolute Percentage Error
    Note: result is NOT multiplied by 100
    """
    result = np.abs((actual - predicted) / (actual + EPSILON))
    return np.mean(np.arctan(np.float32(result)))
maape = maape(actual, forecast)
print('MAAPE: %f' % maape)
Accuracy= 1-maape
print('Accuracy: %f' % Accuracy)
```
